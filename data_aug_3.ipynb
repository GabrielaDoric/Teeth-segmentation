```python
path_to_masks='C:/Users/Gabriela/Desktop/projekt/maske/'
mask_1 = cv2.imread(path_to_masks+"mask_11.jpg")
mask_2 = cv2.imread(path_to_masks+"mask_12.jpg")
mask_3 = cv2.imread(path_to_masks+"mask_48.jpg")
masks = [mask_1, mask_2, mask_3]
```

```python
transform = A.Compose([
  A.RandomScale(scale_limit=0.1, interpolation=1, always_apply=False, p=1)
])
```

```python
transformed = transform(image=image, masks=masks)
transformed_image = transformed['image']
transformed_mask = transformed['masks']
```
