```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import matplotlib
```

```python
mask_path = './apc2016_obj3_json/label.png'
mask = cv2.imread(mask_path)
```

```python
output_path= './apc2016_obj3_json/binary_mask.png'
```

```python
unique_colors = np.unique(mask.reshape(-1, mask.shape[2]), axis=0)
masks = np.zeros(mask.shape)
```

```python
plt.imshow(masks, interpolation='nearest')
plt.show()
```

```python
object_masks = []
for color in unique_colors:
    object_mask = np.where(mask == color, 1, 0)
    object_mask = np.prod(object_mask, axis=-1)
    object_masks.append(object_mask)
    
```

```python
print (len(object_masks))
```

```python
oznake= ['background','11','12','13','14','15','16','17','18',
         '21','22','23','24','25','26','27','28',
         '31','32','33','34','35','36','37','38',
         '41','42','43','44','45','46','47','48',]
for i,mask in enumerate(object_masks):
    plt.imshow(mask, cmap='gray',interpolation='nearest')
    plt.show()
    matplotlib.image.imsave('maske/mask_{}.jpg'.format(oznake[i]), mask, cmap='gray')
```
